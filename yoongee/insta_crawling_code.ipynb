```python
# chromedriver 설치 및 selenium webdriver Import

!apt-get update
!apt install chromium-chromedriver
!pip3 install selenium==3.141


from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

wd = webdriver.Chrome('/Users/yoongee/Desktop/chromedriver',options=options)

from bs4 import BeautifulSoup
import time
import re

# 키워드 검색 함수 정의
def insta_searching(word):
    url = "http://www.instagram.com/explore/tags/" + word
    return url

# 키워드 검색 후 첫번째 게시물 클릭 함수 정의
def select_first(driver):
    first = driver.find_elements_by_css_selector('div._9AhH0')[0]
    first.click()
    time.sleep(3)
    
import re
from bs4 import BeautifulSoup

# 해시태그 크롤링 함수 정의
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    try:
        content = soup.select('div.MOdxS')[0].text
    except:
        content = ''
    tags = re.findall(r'#[^\s#,\\]+', content)
    data = tags
    return data

# 다음 게시물로 넘어가는 함수 정의
def move_next(driver):
    right = driver.find_element_by_css_selector('div.l8mY4.feth3')
    right.click()
    time.sleep(3)
    
# 인스타 id & pw 로그인 
import pandas as pd
df = pd.read_csv('merge_data.csv')
df['insta_tags'] = None

driver = webdriver.Chrome('/Users/yoongee/Desktop/chromedriver')
driver.get('http://www.instagram.com')
time.sleep(3)

email = 'yyk9612@gmail.com'
input_id = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[0]
input_id.clear()
input_id.send_keys(email)

password = 'dbsRl123!'
input_pw = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[1]
input_pw.clear()
input_pw.send_keys(password)
input_pw.submit()

time.sleep(3)

# 식당별 인스타 검색결과 해시태그 크롤링 반복
for k in range (0,365):
    try:
        word = df['rstrt_nm'][k]
        word = str(word)
        url = insta_searching(word)
        driver.get(url)
        time.sleep(8)
        select_first(driver)
        results = []
        target=1
        for i in range(target):
            try:
                data = get_content(driver)
                results = []
                results.append(data)
                move_next(driver)
            except:
                continue
        df['insta_tags'][k] = results[0][:]
    except:
        continue
```
