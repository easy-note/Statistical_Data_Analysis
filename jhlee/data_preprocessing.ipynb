```python
import pandas as pd
csv_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Statistical_Data_Analysis/data/0_origin_data.csv')
```

```python
# for keyword
keywords_list = csv_data.kwrd_list.str.replace(',', '')
csv_data['keywords'] = keywords_list

df = csv_data[['skey', 'rstrt_nm', 'keywords']]
print(df)
```

```python
# for fq_keywords
for i in range(csv_data['kwrd_list'].count()):
  data = csv_data.iloc[i]
  kwrd_list_sp = data["kwrd_list"].split()
  kwrd_list_fq_co_sp = data["kwrd_list_fq_co"].replace(",","").split()
  kwrd = ""
  
  for j in range(len(kwrd_list_sp)):
    kwrd += kwrd_list_sp[j] * int(kwrd_list_fq_co_sp[j])

  csv_data.at[i,'kwrd_all'] = kwrd

keywords_list = csv_data.kwrd_all.str.replace(',', ' ')
csv_data['keywords'] = keywords_list

fq_df = csv_data[['skey', 'rstrt_nm', 'keywords']]

print(fq_df)
```

```python
# for instagram_keywords
csv_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Statistical_Data_Analysis/data/0_instagram_data.tsv',  sep='\t')
csv_data['keywords'] = csv_data['final_keywords'] + ' ' + csv_data['insta_tags']

insta_df = csv_data[['skey', 'rstrt_nm', 'keywords']]
print(insta_df)
```

```python
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Statistical_Data_Analysis/data/keywords.tsv', sep="\t")
fq_df.to_csv('/content/drive/MyDrive/Colab Notebooks/Statistical_Data_Analysis/data/fq_keywords.tsv', sep="\t")
insta_df.to_csv('/content/drive/MyDrive/Colab Notebooks/Statistical_Data_Analysis/data/insta_keywords.tsv', sep="\t")
```
