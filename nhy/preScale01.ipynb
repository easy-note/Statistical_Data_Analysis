```python
df1 = tx1.loc['item'].to_list()
df2 = tx2.loc['item'].to_list()
df3 = tx3.loc['item'].to_list()
df4 = tx4.loc['item'].to_list()
df5 = tx5.loc['item'].to_list()
df6 = tx6.loc['item'].to_list()
```

```python
pd1 = pd.DataFrame(pd.Series(df1[0][0])).T
pd2 = pd.DataFrame(pd.Series(df2[0][0])).T
pd3 = pd.DataFrame(pd.Series(df3[0][0])).T
pd4 = pd.DataFrame(pd.Series(df4[0][0])).T
pd5 = pd.DataFrame(pd.Series(df5[0][0])).T
pd6 = pd.DataFrame(pd.Series(df6[0][0])).T
```
