```python
import pandas as pd
```

```python
insta_csv = pd.read_csv("/content/drive/MyDrive/2022-1_통계기반데이터분석_1팀/data/df2.csv")
final_merge = pd.read_csv("/content/drive/MyDrive/2022-1_통계기반데이터분석_1팀/data/final_merge2.tsv",sep="\t")
```

```python
insta_csv = insta_csv[['skey','rstrt_nm','insta_tags']]
```

```python
insta_csv['insta_tags']
```

```python
target = pd.merge(left = final_merge, right = insta_csv, how = "left", on = ["skey","rstrt_nm"])
target
```

```python
scale_str_arr = []
```

```python
for i in range(len(target)):
  if(type(target['insta_tags'][i])==str):
    scale_str = target['insta_tags'][i].replace('#','').replace("'","").replace(',','').replace('[','').replace(']','')
    print(scale_str)
    scale_str_arr.append(scale_str)
  else:
    scale_str_arr.append('')

scale_str_arr
```

```python
target['insta_tags'] = pd.Series(scale_str_arr) 
```

```python
target = target[['skey','rstrt_nm','final_keywords','insta_tags']]
target
```

```python
target.to_csv("/content/drive/MyDrive/2022-1_통계기반데이터분석_1팀/data/final_merge2.tsv",sep="\t")
```
